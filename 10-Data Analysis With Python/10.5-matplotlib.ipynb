#### Data Visualization With Matplotlib

Matplotlib is a powerful plotting library for Python that enables the creation of static, animates, and interactive visualizations. It is widely used for data visualization in data science and analytics. In this lesson, we will cover the basics of Matplotlib, including creating various types of plots and customizing them.

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

## Create a line plot
plt.plot(x, y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Basic Line Plot")
plt.show()
```

```python
# Create a customized line plot
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, color='red', linestyle='--', marker='o', linewidth=3, markersize=9)
plt.grid(True)
```

```python
## Multiple Plots
# Sample Data
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

plt.figure(figsize=(9, 5))

plt.subplot(2, 2, 1) # 1 row, 2 columns
plt.plot(x, y1, color='green')
plt.title("Plot 1")


plt.subplot(2, 2, 2) # 1 row, 2 columns
plt.plot(y1, x, color='red')
plt.title("Plot 2")

plt.subplot(2, 2, 3) # 1 row, 2 columns
plt.plot(x, y2, color='blue')
plt.title("Plot 3")

plt.subplot(2, 2, 4) # 1 row, 2 columns
plt.plot(x, y2 ,color='red')
plt.title("Plot 4")
```
